Handling Financials by converting to float

```python
def conversion_finance_float(text):
  txt = str(text)
  if txt[0] == '$':
    res = float(txt[1:].replace(',', '.'))
    return res
  return float('nan')

def conversion_financial(df,l):
  dico = {}
  for name in l:
    dico[name] = []
  for name in l:
    for i in range(len(df)):
      dico[name].append(conversion_finance_float(df[name].iloc[i]))
  data = pd.DataFrame(dico, index = df.index)
  df[l] = data[l]

l = ['INCOME', 'HOME_VAL', 'BLUEBOOK', 'OLDCLAIM']
conversion_financial(label_train, l)
conversion_financial(label_test, l)
```

## Mutual information and Pearson correlation

We are facing a classification Problem. We will compute correlation between target and the several features as well as other representation. This serves for linear regression mainly.

```python
# First we compute a correlation : easiest step
corr_ = []
for name in cols_train:
  #print("correlation (target," + name + ")  : ", abs(np.corrcoef(y_train, label_X_train[name])[0,1]))
  corr_.append(abs(np.corrcoef(y_train, label_X_train[name])[0,1]))
idx_ = np.argsort(corr_)
idx_ = idx_[::-1]
correlations = []
features_ = []
for id in idx_:
  correlations.append(corr_[id])
  features_.append(cols_train[id])
fig = plt.figure(figsize = (25,8))
ax = fig.add_axes([0,0,1,1])
ax.set_yticks([0,0.025,0.05,0.075,0.1,0.125,0.15,0.175,0.2, 0.225])
ax.bar(features_, correlations)
ax.set_title("Correlation between features")
plt.grid(axis = 'y')
plt.show()
```

For more general guidance between target and feature dependance we will look into the mutual information

```python
disc_feature = ['KISDRIV', 'HOMEKIDS', 'PARENT1', 'MSTATUS', 'SEX', 'EDUCATION', 'JOB', 'CAR_USE', 'CAR_TYPE', 'RED_CAR', 'CLM_FREQ', 'REVOKED', 'URBANICITY']
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_classif(X, y)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(label_X_train, y_train, disc_feature)


def plot_mi_scores(scores):
    fig = plt.figure(figsize = (10,12))
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores)
```

Let's select features with orders of dependance with target and see how the performance evolves. I will take correlation so it seems legit to also include a logistic reegression to emphasis linear realtionships

```python
def variation_feat_clf(f):
  columns_ = f
  log = []
  tree_ = []
  forest = []
  clf1 = tree.DecisionTreeClassifier()
  clf2 = RandomForestClassifier(bootstrap = True, max_depth = 80, n_estimators = 100)
  clf3 = LogisticRegression(max_iter = 10000)

  features_to_drop = []
  other_features_to_drop = []
  best_tree, best_forest, best_log = 0,0,0

  feat_tree, feat_forest, feat_log = [], [], []
  for i in range(len(columns_)):
    object_cols = columns_[:i+1]
    
    scores = cross_val_score(clf1, label_X_train.loc[:,object_cols], y_train, cv=3)
    scores_forest = cross_val_score(clf2, label_X_train.loc[:,object_cols], y_train, cv=3)
    scores_logistic = cross_val_score(clf3, label_X_train.loc[:,object_cols], y_train, cv=3)
    
    score_tree = round(np.mean(scores)*100,2)
    score_forest = round(np.mean(scores_forest)*100,2)
    score_log = round(np.mean(scores_logistic)*100,2)
    
    tree_.append( score_tree)
    forest.append(score_forest)
    log.append(score_log)
    
    if score_forest > best_forest:
      best_forest = score_forest
      feat_forest = object_cols
    if score_tree > best_tree:
      best_tree = score_tree
      feat_tree = object_cols
    if round(np.mean(scores_logistic)*100,2) > best_log:
      best_log = round(np.mean(scores_logistic)*100,2)
      feat_log = object_cols
        
    if i>0:
      last_score_tree = tree_[i-1]
      last_score_forest = forest[i-1]
      last_score_log = log[i-1]
      if last_score_tree > score_tree and last_score_forest > score_forest and last_score_log > score_log:
        features_to_drop.append(columns_[i])
    
    X_arr = np.arange(len(tree_))

  return log, tree_, forest, features_to_drop, best_tree, best_log, best_forest, feat_tree, feat_forest, feat_log, other_features_to_drop, X_arr
```

```python
log, tree_, forest, features_to_drop, best_tree, best_log, best_forest, feat_tree, feat_forest, feat_log, other_features_to_drop, X_arr = variation_feat_clf(features_)

plt.figure(figsize= (16,8))
plt.plot(X_arr, tree_, label = 'Decision tree')
plt.plot(X_arr, forest, label = 'Random Forest')
plt.plot(X_arr, log, label = 'Logistic')
plt.xlabel("number of features", fontsize = 15)
plt.ylabel("score (%)", fontsize = 15)
plt.grid()
plt.legend()

print("Features to drop : ", features_to_drop)
```
